# This is to filter all the 4 data. We remove the rows with NA values and rows having the same values and save the rest of the rows in a different file.

```python
#Reading the data1 excel file and saving it as a csv file
import pandas as pd
file=pd.read_excel("data1.xlsx")
file.to_csv("data1csv.csv", index=None, header=True)
```

```python
#For analyzing the data1, the data is read into a python variable "csv_reader"
import csv
csv_file=open("data1csv.csv")
csv_reader=list(csv.reader(csv_file))[1:]
```

```python
#To filter the data1
same_val=[]
dummy=csv_reader[:]
print("initial length of the file is", len(csv_reader)) #printing the initial number of rows or attributes in the data
print("The initial lenght of the dummy file is", len(dummy))
for row in csv_reader:
    r=row[1:]
    l=len(r)
    if r.count('NA')>=l-5: #removing the rows that are having less than five values
        dummy.remove(row)
    elif all(ele==r[5] for ele in r):#removing the rows with the same values
        same_val.append(row[0])
        dummy.remove(row)
print("final length of the file is",len(csv_reader)) #printing the number of rows after the filteration
print("The final length of the dummy file is", len(dummy))
with open("filtered_merged_file1.csv", "w") as nfile:
       for row in dummy:
           nfile.write(",".join(row)) #saving the filtered data as a csv file
           nfile.write("\n")
print(len(same_val))
for val in same_val:
    print(val)
```

Analyzing the patient's age in data1 csv file

```python
import csv#importing csv module to work with csv files easily
from matplotlib import pyplot as plt#this module is to plot graphs etc.
import collections
file=open("data1csv.csv")
csvreader=list(csv.reader(file))
for ele in csvreader:
    if ele[0]=="patient.age_at_initial_pathologic_diagnosis":
        age=ele[1:]
        break
#extracting the age row
age.sort()
c=collections.Counter(age)
attribute=[]
frequency=[]
for key, val in c.items():
    attribute.append(key)
    frequency.append(val)
print(c)
fig, ax = plt.subplots(figsize=(14, 4))
ax.tick_params('x', rotation=60)
plt.bar(attribute, frequency)#plotting a bar graph
plt.show()
print(sum(c.values()))
```

# Analyzing Patient's Race

```python
import csv#importing csv module to work with csv files easily
from matplotlib import pyplot as plt#importing pyplot for plotting useful graphs
import collections 
# this module helps us to count the freq of appearance of an element in list
file=open("data1csv.csv")
csvreader=list(csv.reader(file))# reading the csv file as list of rows
for ele in csvreader:
    if ele[0]=="patient.race_list.race":
        print("Yes the patient race is found in the data: ")
        age=ele[1:]
        age.sort()
        c=collections.Counter(age)
        c['NA']=c.pop('')
        attribute=[]
        frequency=[]
        for key, val in c.items():
            attribute.append(key)
            frequency.append(val)
        print(c)
        fig, ax = plt.subplots(figsize=(14, 4))
        plt.bar(attribute, frequency)
        plt.xlabel("Race of the patients")
        plt.ylabel("Number of patients")
        
        plt.show()
        break
```

Since the number of whites is comparitevely very high, the number of american indian or alaska native is not seen properly in the above graph. So a python code is written to apply a break point so that the american indian count is visible in the graph

```python
import pandas as pd
import numpy as np
XX = pd.Series(frequency,index=attribute)
fig, (ax1,ax2) = plt.subplots(2,1,sharex=True,figsize=(24,4))
ax1.spines['bottom'].set_visible(False)
ax1.tick_params(axis='x',which='both',bottom=False)
ax2.spines['top'].set_visible(False)
bs = 300
ts = 600
ax2.set_ylim(0,bs)
ax1.set_ylim(ts,800)
ax1.set_yticks(np.arange(600,800,100))
bars1 = ax1.bar(XX.index, XX.values)
bars2 = ax2.bar(XX.index, XX.values)
plt.xlabel("Race of the patients")
plt.ylabel("Number of patients")
plt.show()
```
